##**Objective 2: Cluster the Customer ( Round 1)**

Our Second Objective is to Segment the Customers using K-Means Clustering, including Standarizing the Data, creating an Inertia Plot, and Interpreting the Clusters.   

#**Scale the Data using Standarization**


Standarize the data so that each column has a Mean of 0 and Standard Deviation of 1.

```python
from sklearn.preprocessing import StandardScaler
```

```python
#scaler = StandardScaler()
#pd.DataFrame(scaler.fit_transform(data_clean)).head()
```

```python
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(data_clean), columns=data_clean.columns)
df_scaled.head()
```

```python
df_scaled.describe().round()
```

##**Fit K-Means Models with 2-15 clusters**

Fit K-Means Clustering Models on the Standarized Data with 2-15 Clusters to create an Inertia Plot.

```python
#Import K-Means and write a loop to fit models with 2-15 Clusters
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

#Create an empty list to hold many Inertia and Silhoutte Values
intertia_values = []
silhoutte_values = []

#Create 2-15 Clusters, and add the Inertia Scores and Silhoute Scores to the list

for k in range(2,16):
  kmeans = KMeans(n_clusters=k,n_init=10, random_state=42) # changed from auto to 10
  kmeans.fit(df_scaled)
  intertia_values.append(kmeans.inertia_)
  silhoutte_values.append(silhouette_score(df_scaled, kmeans.labels_ , metric='euclidean', sample_size=None))
```

```python
#Plot the Inertia Plot
#Turn the list into a Series of Ploting
inertia_series = pd.Series(intertia_values,index=range(2,16))
#Plot the data
inertia_series.plot(marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Number of Clusters vs Inertia')
plt.show()
```

So. looking at this we need to find the **Elbow** of this Inertia Plot

The Inertia drops significantly at k=5. So, we choose k=5 for our KMeans Clustering.

```python
Kmeans5 = KMeans(n_clusters=5, n_init=10, random_state=42)
Kmeans5.fit(df_scaled)
```

##**Check the Number of Customers in each Cluster**

```python
from collections import Counter
Counter(Kmeans5.labels_)
```

#**Create a Heat Map of the Cluster Centers and Interprete the Cluster**

```python
clusters_centers5 = pd.DataFrame(Kmeans5.cluster_centers_, columns=df_scaled.columns)
clusters_centers5.head()
```

```python
plt.figure(figsize=(10,2))
sns.heatmap(clusters_centers5, annot=True,cmap='RdBu',fmt='0.1f',linewidths=0.5)
plt.show()
```

* 0 : Many Products in a Short Time
* 1 : French Customers with Few Products and High Balance
* 2 : German Customers with a High Balance
* 3 : French Customers with More Products and Low Balance
* 4 : Spanish Customers

#**Objective 3: Cluster the Customers (Round 2)**

Our Third Objective is to Segment the Customers using K-Means Clustering using a different Subset of Fields and Compare the Model Results.

##**Updating the Model Dataset**

We will look at the summary stats the Country, and Exclude the Country Field as currently the clusters are dominated by Country Field. Then we use the updated dataset for K-Means Clustering same way as the previous steps.

```python
data_subset.head()
```

```python
data_geo = data_subset.copy()
data_geo.Gender = np.where(data_geo.Gender == 'Female',1,0)
data_geo.head()
```

**Customers from French, Spain and Germany have pretty similar attributes except in Balance. German customoers have higher Balance than the other two countries.**

So, we will exclude the Country Field.

```python
df_scaled.head()
```

```python
df_scaled_no_geo = df_scaled.drop(columns = ['France','Germany','Spain'])
df_scaled_no_geo.head()
```

##**Now Fit the Clustering Model with the Updated Data**

```python
#Import K-Means and write a loop to fit models with 2-15 Clusters
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

#Create an empty list to hold many Inertia and Silhoutte Values
intertia_values = []
silhoutte_values = []

#Create 2-15 Clusters, and add the Inertia Scores and Silhoute Scores to the list

for k in range(2,16):
  kmeans = KMeans(n_clusters=k,n_init=10, random_state=42) # changed from auto to 10
  kmeans.fit(df_scaled_no_geo)
  intertia_values.append(kmeans.inertia_)
  silhoutte_values.append(silhouette_score(df_scaled, kmeans.labels_ , metric='euclidean', sample_size=None))
```

```python
#Plot the Inertia Plot
#Turn the list into a Series of Ploting
inertia_series = pd.Series(intertia_values,index=range(2,16))
#Plot the data
inertia_series.plot(marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Number of Clusters vs Inertia')
plt.show()
```

The Inertia drops significantly at k=4. So, we choose k=4 for our KMeans Clustering.

```python
Kmeans4 = KMeans(n_clusters=4, n_init=10, random_state=42)
Kmeans4.fit(df_scaled_no_geo)
```

##**Check again the Number of Customers in each Cluster**

```python
from collections import Counter
Counter(Kmeans4.labels_)
```

#**Create the Heat Map again to find the Cluster Centers**

```python
clusters_centers4 = pd.DataFrame(Kmeans4.cluster_centers_, columns=df_scaled_no_geo.columns)
clusters_centers4.head()
```

```python
plt.figure(figsize=(10,2))
sns.heatmap(clusters_centers4, annot=True,cmap='RdBu',fmt='0.1f',linewidths=0.5)
plt.show()
```

* 0 : Customers who don't have a Credit Card, but with relatively high tenure.
* 1 : High Balanace with Few Products and Have Credit Card.
* 2 : Low Balance, More Products and Have Credit Card.
* 3 : Customers with Many Products in a Short Period of Time.
